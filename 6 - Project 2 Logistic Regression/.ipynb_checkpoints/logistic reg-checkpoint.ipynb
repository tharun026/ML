```python
#type casting
target = df['Creditability']
df= df.drop(['Creditability'], axis=1)
print(target.head())
```

```python
#to check if there are any missing values
df.isna().any() 
```

```python
from sklearn.linear_model import LogisticRegression
y=target
model = LogisticRegression().fit(df,y)
print(model.score(df,y))
```

```python
#df['Purpose_of_loan'] = df['Purpose_of_loan'].astype(str) 
cat_col = []
num_col = ["Age_in_years","Credit_Amount","Duration_of_Credit_month"]
for col in df.columns:
    if col not in num_col:
        df[col] = df[col].astype(str)
        cat_col.append(col)
        
cat_col 
```

```python
cat_df = df[cat_col]
cat_df.head()
```

```python
dummies = pd.get_dummies(df[cat_col],prefix="cat")
dummies.head()
```

```python
for col in cat_col:
    df = df.drop([col],axis=1)
    
df= pd.concat([df,dummies],axis=1)
```

```python
model = LogisticRegression().fit(df,y)
print(model.score(df,y))
```
